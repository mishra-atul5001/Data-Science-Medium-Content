```python
# Loading Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Dataset
omicron_dataset = pd.read_csv('covid-variants.csv')
omicron_dataset.head()
```

```python
'''
Performing Data-Sanity Checks

1. Unique/Value_Count for each column
2. DTypes for the columns
3. Missing Values/% of Missing Values
4. Identify the Columns for which the DType needs to be changed!
'''

def perform_sanity_check(dataframe):
    # For each column
    for col in dataframe.columns:
        # Print Number of Unique Values and Number of Missing Values
        print('"{}" has {} unique Values and has total number of missing values = {}'.format(col,dataframe[col].nunique(),dataframe[col].isnull().sum()))
        # Print the Top 5 Maxx Values for each column
        print('Top 5 Maxx Values for col {} is\n{}\n'.format(col,dataframe[col].value_counts().iloc[:5]))
    # Let's Check the DType for each column
    print(dataframe.info())

    return

perform_sanity_check(omicron_dataset)
```

```python
# Let's change DType for date
omicron_dataset.date = pd.to_datetime(omicron_dataset.date)
# Derive Month and Year from the date column
omicron_dataset['Month'] = omicron_dataset.date.dt.month_name()
omicron_dataset['Year'] = omicron_dataset.date.dt.year
omicron_dataset.info()
```

```python
perform_sanity_check(omicron_dataset)
```

### Inference 1: Bar-Chart for Top 5 countries with the Most Number of cases  
- Each Year  
- Month observing highest number of cases  

```python
'''
Approach:
- Find count of cases for each country Month wise.
- Put it in a new dataframe
'''

def country_month_analysis(dataframe):
    # Fetch the TOP Countries with highest Cases in the span of 3 years and all months
    top_5_countries = dataframe.groupby(['location']).size().reset_index().rename(columns={0:'num_count_cases'}).sort_values(by='num_count_cases',ascending=False).reset_index(drop=True)[:5].location.values.tolist()
    # Fetch the TOP Months seeing the highest cases
    top_5_months = dataframe.groupby(['Month']).size().reset_index().rename(columns={0:'num_count_cases'}).sort_values(by='num_count_cases',ascending=False).reset_index(drop=True)[:5].Month.values.tolist()
    # Fetch the analysis per location, per year - month wise
    inf_1_df = dataframe.groupby(['location','Year','Month']).size().reset_index().rename(columns={0:'num_count_cases'}).sort_values(by='num_count_cases',ascending=False).reset_index(drop=True)
    return inf_1_df,top_5_countries,top_5_months

inf_1_df,top_5_countries,top_5_months = country_month_analysis(omicron_dataset)
print('Top 5 COVID Outbreak Countries: {}'.format(top_5_countries))
print('Top 5 COVID Outbreak Months: {}'.format(top_5_months))
inf_1_df.head(10)
```

```python
final_list_top_countries = inf_1_df[inf_1_df['location'].isin(top_5_countries)]
final_list_top_countries = final_list_top_countries[final_list_top_countries['Month'].isin(top_5_months)].reset_index(drop=True)
final_list_top_countries.head()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(y= 'location',x = 'num_count_cases',hue = 'Year',data = final_list_top_countries,color = '#35858B');
plt.title('Top 5 Countries outbreak over 2 years!');
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x= 'location',y = 'num_count_cases',hue = 'Month',data = final_list_top_countries,color = '#A267AC');
plt.title('Top 5 Countries outbreak over 2 years based on TOP 5 Months!!');
```

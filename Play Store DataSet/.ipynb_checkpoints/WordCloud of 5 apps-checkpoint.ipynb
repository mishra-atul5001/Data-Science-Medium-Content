What exploration of these categories will lead to, First it will give a way what category should our app belong too. These categories paint a lot clear picture if we look at it. The top category is **Family**, we were pretty shocked to see that **Games** came out to be on second number. 

Now question arises why Family category is number 1, we can answer this question when we see what are the applications are being downloaded in this category.

```python


def remove_plus_add_underscore(apps_data):
    apps_data['Installs'] = apps_data['Installs'].str.replace(',','')
    apps_data['Installs'] = apps_data['Installs'].str.replace('+','')
    apps_data['Content Rating'] = apps_data['Content Rating'].str.replace('+','')
    apps_data['Content Rating'] = apps_data['Content Rating'].str.replace(' ','_')
    apps_data['Installs'] = apps_data['Installs'].astype('int64')
    # Converting SIZE of APPS from KB to MB
    for i,j in enumerate(apps_data.Size):
        #print(i,j)
        if 'k' in j:
            clean_str = j.replace('k','')
            cleant_int = round(float(clean_str)*0.001,2)
        else:
            clean_str = j.replace('M','')
            cleant_int = float(clean_str)
        apps_data.Size.iloc[i] = cleant_int
        #print('Clean',cleant_int)
    return apps_data

apps_data_new = remove_plus_add_underscore(apps_data)
apps_data_new.head()
```

```python
# Let's filter out the data according our top 5 categories 
list_categories =  ['FAMILY', 'GAME', 'TOOLS', 'PERSONALIZATION' ,'LIFESTYLE' ]

df_family = apps_data_new[apps_data_new.Category==list_categories[0]]
df_game = apps_data_new[apps_data_new.Category==list_categories[1]]
df_tools = apps_data_new[apps_data_new.Category==list_categories[2]]
df_person = apps_data_new[apps_data_new.Category==list_categories[3]]
df_lifestyle = apps_data_new[apps_data_new.Category==list_categories[4]]
```

```python
# Now from these top 5 categories let's explore top 5 apps in each category 
# Let's fix some criteria to get our top apps from each category 
# 1. Installs of app should be 1000000 or more 
# 2. Rating of App should be 4.5 or more 
# 3. App Should be free which means price should be ZERO 
# As our condition are set Let's RUMBLE!

def sorted_dataset(df):
    df_sort = df.sort_values(by=['Installs', 'Rating'], ascending=False).head(5)
    return df_sort 
```

```python
# Now we have seen the word distribution, let's create there word clouds 
def create_wordcloud(df):
    comment_words = '' 
    stopwords = set(STOPWORDS) 

    # iterate through the csv file 
    for val in df.Translated_Review: 

        # typecaste each val to string 
        val = str(val) 

        # split the value 
        tokens = val.split() 

        # Converts each token into lowercase 
        for i in range(len(tokens)): 
            tokens[i] = tokens[i].lower() 

        comment_words += " ".join(tokens)+" "

    wordcloud = WordCloud(width = 1000, height = 800, 
                    background_color ='white', 
                    stopwords = stopwords, 
                    min_font_size = 5).generate(comment_words) 

    # plot the WordCloud image                        
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud,interpolation='bilinear') 
    plt.axis("off") 
    plt.tight_layout(pad = 5) 

    plt.show()
```
